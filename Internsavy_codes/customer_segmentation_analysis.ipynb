Importing the libraries which required

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

reading or uploading the data

```python
data = pd.read_csv('/content/Mall_Customers.csv')
```

Segmenting the Customer data by visualizing graph based on different parameters

In below graph, we clearly see that fermales are more than males

```python
sns.countplot(x='Gender',data=data)
plt.title('Distribution of Gender')
```

In thid graph, we clearly see that between 30-40 yr of age have high number of population

```python
data.hist('Age',bins=40)
plt.title('Distribution of age')
plt.xlabel('age')
```

In this graph, we clearly see that at between 30-40 yr of age there is large number of females than males

```python
plt.hist('Age',data = data[data['Gender']=='Male'],alpha=0.5,label='Male')
plt.hist('Age',data = data[data['Gender']=='Female'],alpha=0.5,label='Female')
plt.title('Distribution of age by gender')
plt.xlabel('Age')
plt.legend()
```

In this graph, there is between 60-80k have highest number of counts

```python
data.hist('Annual Income (k$)')
plt.title('Annual Income Distribution in Thousands of Dollars')
plt.xlabel('Thousand of Dollars')
```

In this graph, we clearly see that females have high income

```python
plt.hist('Annual Income (k$)',data = data[data['Gender']=='Male'],alpha=0.5,label='Male')
plt.hist('Annual Income (k$)',data = data[data['Gender']=='Female'],alpha=0.5,label='Female')
plt.title('Distribution of income by gender')
plt.xlabel('Thousand of Dollars')
plt.legend()
```

mean of male and female customers based on their spending

```python
Male_customers = data[data['Gender']=='Male']
Female_customers = data[data['Gender']=='Female']

print(Male_customers['Spending Score (1-100)'].mean())
print(Female_customers['Spending Score (1-100)'].mean())
```

Heatmap for seeing the correlation between various variables:

```python
sns.heatmap(data.corr(),annot=True)
```

Heatmap for female_customers

```python
sns.heatmap(Female_customers.corr(),annot=True)
plt.title('Correlation heatmap - Female')
```

Heatmap for male_customers

```python
sns.heatmap(Male_customers.corr(),annot=True)
plt.title('Correlation heatmap - Male')
```
