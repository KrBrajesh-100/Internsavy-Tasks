import libraries which required

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

read the data

```python
data=pd.read_csv('/content/batsman.csv')
```

```python
data
```

cleaning the data

```python
data.drop(data[data['Bat1']=='DNB'].index,inplace=True)
data
```

converting int type to float

```python
data['SR']=data["SR"].astype(float)
data['Runs']=data["Runs"].astype(float)
data['4s']=data["4s"].astype(float)
data['6s']=data["6s"].astype(float)
```

```python
data.dtypes
```

count top 100 batsman from the data

```python
new_data = data['Batsman'].value_counts()[:100]
new_data
```

display the data of individual batsman

```python
ds_individual = data.query('Batsman=="Shimron Hetmyer"')
ds_individual
```

get the highest number of runs, strike rate, 6's and 4's

```python
print("The highest number of score for this batsman is: ")
print(ds_individual['Runs'].max())
print("The highest Strike Rate for this batsman is: ")
print(ds_individual['SR'].max())
print("The most number of 4s for this batsman is: ")
print(ds_individual['4s'].max())
print("The most number of 6s score for this batsman is: ")
print(ds_individual['6s'].max())
```

get the data against other country's team

```python
againstindia = ds_individual.query('Opposition == "v India"')
againstindia
```

get the runs of batsman that he made on different ground location

```python
plt.figure(figsize=(15,7))
x = againstindia['Ground']
y = againstindia['Runs']
plt.xlabel('Ground Location', labelpad=25)
plt.ylabel('Runs Scored', labelpad=25)
plt.title('Runs vs The Ground', fontweight='bold', pad=30, fontsize=20)
plt.bar(x, y)
```

visualize the data of runs vs strike rate

```python
plt.figure(figsize=(15,7))
x = ds_individual['SR']
y = ds_individual['Runs']
plt.xlabel('Strike Rate', labelpad=30)
plt.ylabel('Runs', labelpad=30)
plt.title('Runs vs Strike Rate', fontweight='bold', pad=20, fontsize=20)
plt.scatter(x, y, color='red', s=300, edgecolor='black')
```

get the data of country vs country match against particular country and get highest run,strike rate,4's and 6's

```python
ds_opposition = data.query('Opposition =="v England"')
ds_opposition
```

```python
print("The highest number of score against this team is: ")
print(ds_opposition['Runs'].max())
print("The highest Strike Rate against this team is: ")
print(ds_opposition['SR'].max())
print("The most number of 4s against this team is: ")
print(ds_opposition['4s'].max())
print("The most number of 6s against this team is: ")
print(ds_opposition['6s'].max())
```
